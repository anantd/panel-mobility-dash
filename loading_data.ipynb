```python
import altair as alt
dat = a.get_state("New York")

heatmap = alt.Chart(
        dat
    ).mark_rect().encode(
        x = alt.X('date(date):O', title = 'day of month'),
        y =alt.Y('month(date):O', title = 'month'),
        color=alt.Color('7_day:Q', scale=alt.Scale(scheme="redyellowgreen"),
                       legend=alt.Legend(title="Driving volume")),
        tooltip=[
            alt.Tooltip('monthdate(date):T', title='Date'),
            alt.Tooltip('driving_7_day:Q', title='Requests vol', format = '.0f')
        ]
    ).properties(
        width=500,
        height = 200,
        title = {'text': f'Driving directions requests volume', 
                'subtitle': '7-day rolling average, indexed to Jan 13 requests',
                'fontSize': 16,
                'subtitleFontSize': 10}
    )


heatmap
```

```python
import pandas as pd
import altair as alt
s1 = g.get_state("Maryland")
s2 = g.get_state("Washington")
goog = pd.concat([s1, s2])

work = pd.DataFrame(goog[goog.destination_type == "workplaces"])
grocery = pd.DataFrame(goog[goog.destination_type == "grocery_pharmacy"])
park = pd.DataFrame(goog[goog.destination_type == "parks"])

#park = 

wk = alt.Chart(work).mark_line(interpolate = 'monotone').encode(
    x = 'date:T',
    y = 'volume:Q',
    color = 'sub_region_1:N'
).properties(title = 'Workplace traffic')


gr = alt.Chart(grocery).mark_line(interpolate = 'monotone').encode(
    x = 'date:T',
    y = 'volume:Q',
    color = 'sub_region_1:N'
).properties(title = 'Grocery/pharmacy traffic')

pk = alt.Chart(park).mark_line(interpolate = 'monotone').encode(
    x = 'date:T',
    y = 'volume:Q',
    color = 'sub_region_1:N'
).properties(title = 'Parks traffic')


wk | gr | pk
```
